```python
import numpy as np
import pandas as pd
import json
```

# Loading evidence from json

```python
with open("../data/raw/release_evidences.json") as file:
	data = json.load(file)
```

```python
B_type_data = {}
for i in data:
    if data[i]['data_type'] == "B":
        B_type_data[i] = data[i]
```

```python
B_type_data_colnames = []
for i in B_type_data:
    B_type_data_colnames.append(i)
```

# Train

```python
df = pd.read_csv("../data/raw/release_train_patients.csv")
df.head()

all_clear_rows = []
for i in df['EVIDENCES']:
	row = i.split("', '")
	row[-1] = row[-1][:-2]
	row[0] = row[0][2:]
	clear_row = []
	for i in row:
		if "@" not in i:
			clear_row.append(i)
	all_clear_rows.append(clear_row)

df['EVIDENCE_clear'] = all_clear_rows

zeros = pd.DataFrame(np.zeros((len(all_clear_rows), 208)).astype(int), columns=B_type_data_colnames)
for i in range(len(all_clear_rows)):
    zeros.loc[i, all_clear_rows[i]] = 1
    
zeros['SEX'] = (df['SEX'] == "M").astype(int)
zeros['PATHOLOGY'] = df['PATHOLOGY'] 
zeros.to_csv("../data/processed/train.csv")

zeros.head()
```
